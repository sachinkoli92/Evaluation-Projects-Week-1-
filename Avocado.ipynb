```python
# Removing skewness using Log Transform

for i in df.columns:
    if df.skew().loc[i]>0.55:
        df[i]=np.log1p(df[i])
```

```python
# Removing Outliers
```

```python
from scipy.stats import zscore
```

```python
z=np.abs(zscore(df))
```

```python
threshold=3
df1=df[(z<3).all(axis=1)]
print(df.shape)
print('\n')
print('Removing Ouliers, Now shape is:')
print(df1.shape)
```

```python
# Separating the target variable- Average Price
```

```python
x=df1.drop(['AveragePrice'],axis=1)
x
```

```python
y=df1['AveragePrice']
```

```python
y
```

```python
#Applying Standard Scaler to dependent Variable.
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
xx=sc.fit_transform(x)
xx=pd.DataFrame(xx,columns=x.columns)
```

```python
xx
```

```python
xx.skew()
```

```python
x=xx
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso,Ridge,ElasticNet
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.25,random_state=42)
```

```python
print('Shape of X train: ' , x_train.shape)
print('Shape of Y train: ' , y_train.shape)
print('Shape of X test: ' , x_test.shape)
print('Shape of Y test: ' , y_test.shape)
```

```python
list=[LinearRegression(),Lasso(),Ridge(),ElasticNet(),DecisionTreeRegressor(),SVR(),KNeighborsRegressor()]
for i in list:
    i.fit(x_train,y_train)
    predict=i.predict(x_test)
    print('Score of: ',i, 'is: ', i.score(x_train,y_train))
    print('R2 Score corresponding to: ',i,'is: ',r2_score(y_test,predict))
    print('Mean Absolute Error: ', mean_absolute_error(y_test,predict))
    print('Mean Squared Error: ',mean_squared_error(y_test,predict))
    print('Root Mean Squared Error: ',np.sqrt(mean_squared_error(y_test,predict)))
    print('\n')
    print('★★★★★★★★★★★★★★★★★★★★★★★★★★★★★★★★★★★★★★★★★★★★★★★★★★')
    print('\n')
```

```python
# Using ensemble technique to boost our score

import numpy as np
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor(n_estimators=20,random_state=42)
rf.fit(x_train,y_train)
print('Score of random forest is:',rf.score(x_train,y_train))
predict=rf.predict(x_test)
print('Mean absolute error: ',mean_absolute_error(y_test,predict))
print('Mean squared error: ',mean_squared_error(y_test,predict))
print('Root mean squared error:',np.sqrt(mean_squared_error(y_test,predict)))
print('R2 score: ',r2_score(y_test,predict))
```

```python
from sklearn.model_selection import GridSearchCV
rf=RandomForestRegressor()
parameters={'criterion':['mse','mae'],'random_state':range(42,60)}
cv=GridSearchCV(rf,parameters)
cv.fit(x,y)
cv.best_params_
```

```python
#Cross Validation for random forest regressor
from sklearn.model_selection import cross_val_score
rf=RandomForestRegressor(criterion='mae',random_state=54)
score=cross_val_score(rf,x,y,cv=3,scoring='r2')
print('Score: ',score)
print('Mean Score: ',score.mean())
print('Standard Deviation: ',score.std())
```

Working on the best model i.e, Random forest regressor as it has the highest r2 score and the errors are low. Also it does not overfit easily.

```python
rfr=RandomForestRegressor(criterion='mae',n_estimators=20,random_state=54)
rfr.fit(x_train,y_train)
print('Score of random forest is:',rfr.score(x_train,y_train))
predict=rfr.predict(x_test)
print('Mean absolute error: ',mean_absolute_error(y_test,predict))
print('Mean squared error: ',mean_squared_error(y_test,predict))
print('Root mean squared error:',np.sqrt(mean_squared_error(y_test,predict)))
print('R2 score: ',r2_score(y_test,predict))
```

```python
#Saving the model-Random forest regressor
import joblib
joblib.dump(rfr,'avocado.pkl')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
g=[DecisionTreeClassifier(),SVC(),KNeighborsClassifier(),RandomForestClassifier()]
for j in g:
    max_r_score=0
    print('for the algorithm :' ,j)
    for i in range(33,101):
        x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=i,test_size=.25)
        j.fit(x_train,y_train)
        pred=j.predict(x_test)
        score=j.score(x_train,y_train)
        accuracy=accuracy_score(pred,y_test)
        print('accuracy score corresponding to random state',i,'is: ',accuracy)
        if accuracy>max_r_score:
            max_r_score=accuracy
            final=i
    print('\n')
    print('Max accuracy score corresponding to: ', final,'is: ',max_r_score)
    print('\n')
        
```

```python
# Grid search for random forest Classifier (as we see from all above models random forest classifier accuracy score is best)

from sklearn.model_selection import GridSearchCV
rfc=RandomForestClassifier()
parameters={'criterion':['gini','entropy'],'random_state':range(65,80)}
cv=GridSearchCV(rfc,parameters)
cv.fit(x,y)
cv.best_params_
```

```python
rfc=RandomForestClassifier(criterion='entropy',random_state=78)
rfc.fit(x_train,y_train)
predict=rfc.predict(x_test)
print('Score of Random forest classifier is: ',rfc.score(x_train,y_train))
print('Accuracy Score is: ',accuracy_score(y_test,predict))
print('Classification Matrix :',classification_report(y_test,predict))
print('confusion matrix: ',confusion_matrix(y_test,predict))
print('⭐⭐⭐⭐⭐⭐⭐⭐⭐⭐⭐⭐⭐⭐⭐⭐⭐⭐⭐⭐⭐⭐⭐⭐⭐⭐⭐')
```

```python
#Cross Validation for Random Forest Regressor
score=cross_val_score(rfc,x,y,cv=5,scoring='accuracy')
print('Score: ',score)
print('Mean Score: ',score.mean())
print('Standard Deviation: ',score.std())
```
